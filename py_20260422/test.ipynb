```python
import xlwings as xw

app = xw.App(add_book = False)
app
```

```python
#엑셀 워크북
wb = app.books.add()
wb
```

```python
# 워크시트 선택 (이름)
ws = wb.sheets['Sheet1']
ws
```

```python
# 시트 이름 변경
ws.name = "영업1팀"
```

```python
# 다른 이름으로 엑셀 저장
wb.save('교육이수현황.xlsx')
```

```python
#워크북 닫기
wb.close()
```

```python
#엑셀 앱 닫기
app.quit()
```

```python
app = xw.App(add_book = False)
```

```python
# 엑셀 워크북 불러오기
wb = app.books.open('교육이수현황.xlsx')
```

```python
wb.sheets.add('영업 2팀')
wb.sheets.add('마케팅팀')
```

```python
wb.sheets.add(name = '세무회계팀', before = '영업 2팀')
```

```python
wb.sheets.add(name = '영업 3팀', before = '영업 2팀')
```

```python
# 워크 시트 선택 (인덱스)
wb.sheets[3]
```

```python
# 시트 삭제
wb.sheets['영업 3팀'].delete()
```

```python
# 시트 활성화
wb.sheets['마케팅팀'].activate()
```

```python
# 엑셀 저장
wb.save()
```

```python
# 엑셀 앱 닫기
app.quit()
```
